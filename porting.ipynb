```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
df
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df
```

```python
movie_df = pd.read_parquet('movie_df.parquet')
movie_df
```

```python
new_df = pd.DataFrame(columns=movie_df.columns)
```

```python
new_df['Title'] = df['name']
new_df['Watched?'] = pd.Series(['Yes' if e else 'No' for e in df['w']])
new_df['H'] = df['h']
new_df['M'] = df['m']
```

```python
import datetime as dt
new_df['Date Added'] = dt.datetime.now().strftime("%m/%d/%Y, %H:%M:%S")
new_df
```

```python
new_df.to_parquet('movie_df.parquet')
```
